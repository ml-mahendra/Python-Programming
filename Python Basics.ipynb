## Sequences
    -String
    -List
    -Tuple

```python
# String: A group of characters enclosed within single or double quotation marks. (Ex: 'M', "Hello")

my_name="Mahendra"
print(my_name)
type(my_name)
```

```python
# List: List is a collection data of different data types. (Ex: list_bio=["name",20,13.5]. 
#List is mutable, can be changed or modified once after declared and initialized.
# Indexing in List starts from 0. 

lst=["Mahendra",21,5.7,"Bangalore"] #Declared and Initialized List

print(lst) # Accessing the list

#Accessing the list with index
print(lst[0])
print(lst[1])
print(lst[2])
print(lst[3])
```

```python
# Since List is mutable, which means we can alter the values. lets change some values

lst[0] = "Mahendra Nath Reddy" # assining the new value to the index 0

print(lst) 

print(len(lst)) #len() will give the total number of elements in the list
```

```python
# Tuple: Tuples is also a collection of different data types. (Ex: tp=(5,"maze",9.7). 
#Tuples are immutable, cannot be changed or modified once declared. 

tp=(5,"Maze",9.7)

print(tp) 

#Accessing elements using index
print(tp[0])
print(tp[1])
print(tp[2])

print(tp[3]) # Since we dont have index 3 it will throw an error, index out of range.
```

```python
# Since tuple is immutable, which means we can not alter the values. lets try to change values

print(len(lst)) #len() will give the total number of elements in the list

tp[0] = 7 # assining the new value to the index 1

print(tp)  # throws error since tuple is immutable
```

```python
# Sets: It is an unordered collection of elements. we can't get the elements of a particular index in sets. 

s={0,1,2,3,4,7}

print(s)

print(s[0]) # trying to print the elements, with index. But, throws error: set object doesn't support indexing. 
```

```python
list1=[10,5,6,7] 
set_a = set(list1) # converting a list to set
print(set_a)
```

```python
# Dictionary: It stores the data in Key:value pair formate. Word:Meaning :: Key:Value

Dic={10:"Oct",12:"Dec"}

#since its key:value pair, index wont work. we use key to extract the values.

print(Dic[10])
print(Dic[12])
```

### Range Function

```python
# Range() will print the sequence, in a particular range.
# Syntax: range(start,stop,step) stop value is excluded, if (range(1,5) then 5 is excluded. it will print seq from 0 to 4 which are 5 elements)
# we will use typecasting to list so that we can see all the elements within the range given in a list form.
r=list(range(1,10))
print(r) # prints 0 to 9

r1=list(range(1,11,2)) 
print(r1) # prints odd numbers till 11 excluding 11. 
```

## Operators

General Purpose of an program is to accept some data and perform some operation and generate the result. While performing operations we may need to do calculation or comparision or condition etc. 

Any symbols that perform meaningfull operation is known as operator. 

### Operators
    -Arithmetic Operators
    -Assignment Operators
    -Relational Operators
    -Logical Operators
    -Boolean Operators

### Arithmetic Operators



```python
# Examples

a=10
b=7

print(a+b) # add
print(a-b) # Sub
print(a*b) # Multiply
print(a/b) # Divide
print(a%b) # Reminder Division
print(a//b) # floor division, gives floor value of Quotient
print(a**b) # power operator
```

### Assignment Operator


```python
# Examples of Assignment Operators
a = 10

b = a
print(b)

# Add and assign value
b += a
print(b)

# Subtract and assign value
b -= a
print(b)

# multiply and assign
b *= a
print(b)

# bitwise lishift operator
b <<= a
print(b)
```

### Relational Operators


```python
x = 10
y = 12

# Output: x > y is False
print('x > y is',x>y)

# Output: x < y is True
print('x < y is',x<y)

# Output: x == y is False
print('x == y is',x==y)

# Output: x != y is True
print('x != y is',x!=y)

# Output: x >= y is False
print('x >= y is',x>=y)

# Output: x <= y is True
print('x <= y is',x<=y)
```

### Logical Operators


```python
# Examples of Logical Operator
a = True
b = False

# Print a and b is False
print(a and b)

# Print a or b is True
print(a or b)

# Print not a is False
print(not a)
```

### Boolean Operators

```python
# Combination of Logical and Assignment

a=True
b=False
print(a and a)
print(a and b)
print(a or b)
print(a or a)
print(not a)
```
