## Loop Types

    - while loop
    - for loop

## Loop Control statements

    - break statement
    - continue statement
    - pass statement



### A while loop statement in Python programming language repeatedly executes a target statement as long as a given condition is true. The loop stops running when the condition fails (become false), and the execution will move to the next line of code.
            Syntax: while expression:
                        statements


```python
x=0
while x<6:
    print(x)
    x=x+1
```

### It has the ability to iterate over the items of any sequence, such as a list or a string. Loop statements will be executed for each item of the sequence.
        Syntax: for iterating_var in Sequence:
                        statements(s)


```python
list = [1, 2, 3, 4, 5]
for num in list:
    #print(list) #prints the list
    print(num) #prints the elements in the list
```

```python
# if you use for loop over an dictionary, it gives us only the keys as output. 

di={"name":"Mahendra","age":21,"University":"REVA"} # creating the dictionary

for ele in di:
    print(ele) # trying to print the elements in the dict. But it gives us only the key's. 
# in order to get the values of the dict, we must use values()

for i in di.values():
    print(i)
    
# TO get both keys and values together, use items()

for a,b in di.items():
    print(a, " ",b)
```

### Python provides us with 3 types of Control Statements:

        - Continue
        - Break
        - Pass

### Continue Statement:

    When the program encounters a continue statement, it will skip the statements which are present after the continue statement inside the loop and proceed with the next iterations. 
    Used to skip one iteration from loop

```python
a=[0,1,2,3,4,5]
for a in a:
    if a ==3:
        continue
    print(a)
    
# Whenever i is gonna be 3, all the steps inside for loop will be skiped for that iteration
```

```python
i=0
while i<=5:
    if i==4:
        continue
    print(i)
    i=i+1
# since, it sjips all the steps in while loo[, when i becomes 4 and it increase i values and loops becomes infinite and runs forever.
```

```python
# if we add increment before continue we will get expected output

i=0
while i<5:
    i=i+1
    if i==4:
        continue
    print(i)
   
```

### Break Statement:

The break statement is used to terminate the loop containing it, the control of the program will come out of that loop.

```python
a=[6,7,8,9,10,11]
for b in a:
    print(b)
```

```python
# But i need to stop my loop when i find 9 

for c in a:
    if c == 9:
        break
    print(c)
```

```python
z=0
while z<8:
    if z == 7:
        break
    print(z)
    z=z+1
```
